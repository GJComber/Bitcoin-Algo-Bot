## Creating a trading algorithm using EMA technical indicators

```python
# Create a DataFrame with the index and Close column from the dataset
ema_signals_df = df.loc[:, ["close"]].copy()

# Set the short window and long windows
short_window = 50
long_window = 100

# Generate the short and long moving averages (50 and 100 days, respectively)
ema_signals_df["EMA50"] = df["close"].ewm(span=short_window).mean()
ema_signals_df["EMA100"] = df["close"].ewm(span=long_window).mean()

# Review the DataFrame
ema_signals_df.head()
```

```python
# Set the Signal column
ema_signals_df["Signal"] = 0.0

# Generate the trading signal 1 or 0,
# where 1 is when the EMA50 is greater than (or crosses over) the EMA100
# where 0 is when the EMA50 is under the EMA100
ema_signals_df["Signal"][short_window:] = np.where(
    ema_signals_df["EMA50"][short_window:] > ema_signals_df["EMA100"][short_window:], 1.0, 0.0
)

# Calculate the points in time at which a position should be taken, 1 or -1
ema_signals_df["Entry/Exit"] = ema_signals_df["Signal"].diff()

# Review the DataFrame
ema_signals_df.head()
```

```python
# Visualize entry position relative to close price
entry = ema_signals_df[ema_signals_df["Entry/Exit"] == 1.0]["close"].hvplot.scatter(
    color='purple',
    marker='^',
    size=200,
    legend=False,
    ylabel='Price in $',
    width=1000,
    height=400
)

# Visualize exit position relative to close price
exit = ema_signals_df[ema_signals_df["Entry/Exit"] == -1.0]["close"].hvplot.scatter(
    color='orange',
    marker='v',
    size=200,
    legend=False,
    ylabel='Price in $',
    width=1000,
    height=400
)

# Visualize close price for the investment
security_close = ema_signals_df[["close"]].hvplot(
    line_color='lightgray',
    ylabel='Price in $',
    width=1000,
    height=400
)

# Visualize moving averages
moving_avgs = ema_signals_df[["EMA50", "EMA100"]].hvplot(
    ylabel='Price in $',
    width=1000,
    height=400
)

# Overlay plots
entry_exit_plot = security_close * moving_avgs * entry * exit
entry_exit_plot
```

## Creating Machine Learning Algorithmic by using SMA Strategy

## Data Prep for Machine Learning Model 

```python
df_btc.head()
```

```python
df_ml = df_btc.drop(columns = ["exchange","trade_count","vwap","symbol"])
```

```python
df_ml
```

```python
# Calculate the daily returns using the closing prices and the pct_change function
df_ml["actual return"] = df_ml["close"].pct_change()

df_ml.head()
```

```python
# Drop all Nan Values from DataFrame
df_ml = df_ml.dropna()


display(df_ml.head())
display(df_ml.tail())
```

## Generating the Features and Target Sets

### Creating the Features Set

```python
# Define a window size of 7
short_window = 7


# Create an SMA that uses short_window, and assign it to a new column named “sma_fast”
df_ml["sma_fast"] = df_ml["close"].rolling(window=short_window).mean()
```

```python
# Define a window size of 80
long_window = 80

# Create an SMA that uses long_window, and assign it to a new columns named “sma_slow”
df_ml["sma_slow"] = df_ml["close"].rolling(window=long_window).mean()
```

```python
# Drop the NaNs using dropna()
df_ml = df_ml.dropna()
```

```python
# Assign a copy of the `sma_fast` and `sma_slow` columns to a new DataFrame called `X`
X = df_ml[["sma_fast", "sma_slow"]].shift().dropna().copy()

# Display sample data
display(X.head())
display(X.tail())
```

### Creating the Target Set

```python
# Create a new column in the `trading_df` called "signal" setting its value to zero.
df_ml["signal"] = 0.0
```

```python
# Create the signal to buy
df_ml.loc[(df_ml["actual return"] >= 0), "signal"] = 1
```

```python
# Create the signal to sell
df_ml.loc[(df_ml["actual return"] < 0), "signal"] = -1
```

```python
# Copy the new "signal" column to a new Series called `y`.
y = df_ml["signal"].copy()
```

## Split the Data into Training and Testing Sets

```python
# Import required libraries
from pandas.tseries.offsets import DateOffset
```

```python
# Select the start of the training period
training_begin = X.index.min()

# Display the training begin date
print(training_begin)
```

```python
# Select the ending period for the training data with an offset of 18 months
training_end = X.index.min() + DateOffset(months=18)

# Display the training end date
print(training_end)
```

```python
# Generate the X_train and y_train DataFrames
X_train = X.loc[training_begin:training_end]
y_train = y.loc[training_begin:training_end]
```

```python
# Generate the X_test and y_test DataFrames
X_test = X.loc[training_end:]
y_test = y.loc[training_end:]
```

```python
# Import required libraries
from sklearn.preprocessing import StandardScaler
```

## Standardize the Data

```python
# Create a StandardScaler instance
scaler = StandardScaler()
 
# Apply the scaler model to fit the X-train data
X_scaler = scaler.fit(X_train)
 
# Transform the X_train and X_test DataFrames using the X_scaler
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

### Incorporate the SVM Machine Learning Into the Trading Strategy

```python
# Import the SVM model
from sklearn import svm
from sklearn.metrics import classification_report
```

```python
# Create the classifier model
svm_model = svm.SVC()

# Fit the model to the data using X_train_scaled and y_train
svm_model = svm_model.fit(X_train_scaled, y_train)
```

#### Backtesting a Machine Learning Trading Algorithm

```python
# Use the trained model to predict the trading signals for the testing data.
svm_test_predictions = svm_model.predict(X_test_scaled)
```

```python
# Evaluate the model's ability to predict the trading signal for the testing data
testing_report = classification_report(y_test, svm_test_predictions)

# Display the report
print(testing_report)
```

```python
# Create a predictions DataFrame
predictions_df = pd.DataFrame(index=X_test.index)

predictions_df["predicted_signal"] = svm_test_predictions

predictions_df["actual return"] = df_ml["actual return"]

predictions_df["SVM_algo_returns"] = (
    predictions_df["actual return"] * predictions_df["predicted_signal"]
)

# Review the DataFrame
predictions_df.head()
```
